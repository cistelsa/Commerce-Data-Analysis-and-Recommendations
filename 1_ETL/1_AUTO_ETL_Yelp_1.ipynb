```python
import pandas as pd
import numpy as np
import gdown
```

```python
# Función para extraer el dataset y guardar una copia
def extract_datsets(id_google, file_name):
    # Enlace directo al archivo en Google Drive (asegúrate de que sea el ID correcto del enlace de descarga)
    dataset_g_drive = f"https://drive.google.com/uc?id={id_google}"
    response = gdown.download(dataset_g_drive, output=f"/lakehouse/default/Files/data/original/Yelp/{file_name}", quiet=True)
```

```python
# Descarga dataset "user.parquet"
id_google = "1TT4ARRIV6i2fO1b5yb0aSUkjhxMb9u6g"
file_name = "user.parquet"

extract_datsets(id_google, file_name)
```
